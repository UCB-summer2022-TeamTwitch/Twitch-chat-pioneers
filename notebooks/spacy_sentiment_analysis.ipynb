```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')
```

```python
text_df = pd.read_csv('../data/Fact Tables/EmoteFactTable.csv')
```

```python
text_df
```

```python
list_polarity = []
for e, t in zip(text_df['Emote'], text_df['Definition 1']):
    doc = nlp(t)
    # print(doc._.blob.polarity)
    list_polarity.append(doc._.blob.polarity)
    
text_df['spacy_polarity'] = list_polarity
```

```python
text_df
```

```python
text_df.to_csv('../data/Fact tables/EmoteFactTable_spacy.csv')
```
